```python
def weighted_categorical_distribution(data, group, var, weight):
    # Somme des poids par catégorie
    grouped = data.groupby([group, var], as_index=False)[weight].sum()
    
    # Calcul du total des poids par groupe
    total = grouped.groupby(group, as_index=False)[weight].sum().rename(columns={weight: 'total_weight'})
    
    # Fusion pour calculer les pourcentages
    merged = pd.merge(grouped, total, on=group)
    merged[f'%_{var}'] = 100 * merged[weight] / merged['total_weight']
    
    # On garde uniquement les colonnes utiles
    return merged[[group, var, f'%_{var}']]
```

```python
# Liste des variables à agréger
cat_vars = ['sexe', 'resid', 'religion']

# Calcul de la répartition pondérée pour chaque variable
repartitions = []
for var in cat_vars:
    temp = weighted_categorical_distribution(data_ehcvm, 'departement', var, 'hhweight')
    repartitions.append(temp)

# Fusion des résultats
final = repartitions[0]
for df in repartitions[1:]:
    final = pd.merge(final, df, on='departement', how='outer')


```

```python
# On crée trois tables pivotées pour chaque variable
pivot_sexe = final.pivot_table(index='departement', 
                            columns='sexe', 
                            values='%_sexe', 
                            aggfunc='mean').add_prefix('%_sexe_')

pivot_resid = final.pivot_table(index='departement', 
                             columns='resid', 
                             values='%_resid', 
                             aggfunc='mean').add_prefix('%_resid_')

pivot_religion = final.pivot_table(index='departement', 
                                columns='religion', 
                                values='%_religion', 
                                aggfunc='mean').add_prefix('%_religion_')

# Fusionner les trois tables
df_final = pivot_sexe.join(pivot_resid, how='outer').join(pivot_religion, how='outer')

# Remettre l'index comme colonne
df_final.reset_index(inplace=True)

# Afficher le résultat

```

```python


def weighted_binary_indicators(data, group, weight, binary_vars):
    """
    Calcule la proportion pondérée (%) de 1 pour chaque variable binaire, par groupe.
    """
    results = []

    for var in binary_vars:
        if var not in data.columns:
            print(f" Variable '{var}' non trouvée, ignorée.")
            continue

        temp = (
            data.groupby(group)
            .apply(lambda x: 100 * (x[var] * x[weight]).sum() / x[weight].sum())
            .reset_index(name=f'{var}_pct')
        )
        results.append(temp)

    # Fusionner toutes les variables sur 'departement'
    final = results[0]
    for t in results[1:]:
        final = pd.merge(final, t, on=group, how='outer')

    return final
```

```python
binary_vars = ['mur', 'toit', 'sol','scol','telpor','internet','elec_ac','eauboi_ss']

agg_binary = weighted_binary_indicators(
    data=data_ehcvm,
    group='departement',
    weight='hhweight',
    binary_vars=binary_vars
)

print(agg_binary.head())
```

```python
import pandas as pd

def weighted_numeric_sums(data, group, weight, numeric_vars):
    """
    Calcule la somme pondérée pour chaque variable numérique, par groupe.
    """
    results = []

    for var in numeric_vars:
        if var not in data.columns:
            print(f"Variable '{var}' non trouvée, ignorée.")
            continue

        temp = (
            data.groupby(group)
            .apply(lambda x: (x[var] * x[weight]).sum())
            .reset_index(name=f'{var}_sum')
        )
        results.append(temp)

    # Fusionner toutes les variables sur le groupe
    final = results[0]
    for t in results[1:]:
        final = pd.merge(final, t, on=group, how='outer')

    return final

```

```python
# Exemple d'utilisation
numeric_vars = ['volail', 'petitrum', 'grosrum']  # tes variables numériques
agg_numeric = weighted_numeric_sums(
    data=data_ehcvm,
    group='departement',
    weight='hhweight',
    numeric_vars=numeric_vars
)

print(agg_numeric.head())
```

```python
# Liste des colonnes existantes
cols = ['departement', '%_sexe_Féminin', '%_sexe_Masculin', '%_resid_Non',
       '%_resid_Oui', '%_religion_Animiste', '%_religion_Autre Réligion',
       '%_religion_Chrétien', '%_religion_Musulman',
       '%_religion_Sans Réligion', 'mur_pct', 'toit_pct', 'sol_pct',
       'scol_pct', 'telpor_pct', 'internet_pct', 'elec_ac_pct',
       'eauboi_ss_pct', 'volail_sum', 'petitrum_sum', 'grosrum_sum']

# Fonction pour nettoyer les noms
def clean_column_name(col):
    col = col.lower()  # tout en minuscules
    col = col.replace('%_', '')  # enlever %_
    col = col.replace(' ', '_')  # remplacer espace par _
    col = col.replace('é', 'e').replace('è', 'e')  # enlever accents
    col = col.replace('ç', 'c')
    col = col.replace('à', 'a')
    col = col.replace('â', 'a')
    return col

# Appliquer à toutes les colonnes sauf 'departement'
new_cols = ['departement'] + [clean_column_name(c) for c in cols if c != 'departement']

# Renommer dans le DataFrame
ehcvm_agg.columns = new_cols

print(ehcvm_agg.head())
```
