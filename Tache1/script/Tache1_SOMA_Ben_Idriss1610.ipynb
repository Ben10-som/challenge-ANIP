```python
import pandas as pd
import matplotlib.pyplot as plt

# Colonnes d'ann√©es
years = list(range(1989, 2031))
years_str = [str(y) for y in years]  # nom des colonnes dans le DataFrame

# Filtrer indicateurs GDP
gdp_total = data_economic[data_economic['INDICATOR'] == 
                          "Gross domestic product (GDP), Constant prices, Percent change"]
gdp_per_capita = data_economic[data_economic['INDICATOR'] == 
                               "Gross domestic product (GDP), Constant prices, Per capita, purchasing power parity (PPP) international dollar, ICP benchmark 2021"]

# Ne garder que les colonnes des ann√©es et convertir en float
gdp_total_values = gdp_total[years_str].astype(float).iloc[0]
gdp_per_capita_values = gdp_per_capita[years_str].astype(float).iloc[0]

# Graphiques
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(years, gdp_total_values, marker='o')
plt.title("PIB total (prix constants, % de variation)")
plt.xlabel("Ann√©e")
plt.ylabel("% de variation")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(years, gdp_per_capita_values, marker='o', color='orange')
plt.title("PIB par habitant (prix constants, PPA $ 2021)")
plt.xlabel("Ann√©e")
plt.ylabel("USD")
plt.grid(True)

plt.tight_layout()
plt.show()
```
